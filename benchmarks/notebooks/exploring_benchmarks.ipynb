```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
run_path = '../runs'
test_name = 'HarmonicOscillator1D_LowFreq'
# test_name = 'Heat_Eq_1+1D'

test_path = os.path.join(run_path, test_name)
```

```python
runs = sorted(os.listdir(test_path))
runs
```

```python
def plot_model(model):
    model_mse = np.load(os.path.join(test_path, model, 'test_mse.npz'))['mse']
    plt.semilogy(model_mse, label=model)

def plot_models(models):
    plt.figure(figsize=(10, 5))

    for model in models:
        plot_model(model)

    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.title(test_name)
    plt.legend()
    plt.show()

# Plot all models
plot_models(runs)
```

```python
non_adaptive_models = [m for m in runs if '_A_' not in m]
plot_models(non_adaptive_models)
```

```python
adaptive_models = [m for m in runs if '_A_' in m]
plot_models(adaptive_models)
```

```python
non_adaptive_ckans = [m for m in non_adaptive_models if 'CKAN' in m]
plot_models(non_adaptive_ckans)
# CKAN_8 is best
```

```python
non_adaptive_lkans = [m for m in non_adaptive_models if 'LKAN' in m]

plot_models(non_adaptive_lkans)
# CKAN_8 is best
```

```python
d = ['FCN_8', 'FCN_A_8', 'LKAN_16', 'CKAN_8', 'CKAN_A_8', 'LKAN_A_8']
plot_models(d)
```

```python
fcns = [m for m in runs if 'FCN_A_' in m]
plot_models(fcns)
```

```python
ckanas = [m for m in runs if 'CKAN_A_' in m]
plot_models(ckanas)
```
