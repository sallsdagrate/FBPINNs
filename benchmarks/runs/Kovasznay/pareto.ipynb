```python
s=0
rE = lambda exact, pred: ((exact - pred) / (np.abs(exact)))
re_ckan = rE(exact0[1:199, 1:199, s], pred0_ckan[1:199, 1:199, s])
re_fcn = rE(exact0[1:199, 1:199, s], pred0_fcn[1:199, 1:199, s])
print(np.mean(re_ckan), np.mean(re_fcn), np.std(re_ckan), np.std(re_fcn))

plt.hist(re_ckan.flatten(), bins=300, label='CKAN')
plt.hist(re_fcn.flatten(), bins=300, label='FCN', alpha=0.5)
plt.legend()
plt.xlabel('Relative Error')
plt.ylabel('Frequency')
plt.show()
```
