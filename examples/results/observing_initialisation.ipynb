```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('../exact_0.npy', 'rb') as f:
    e0 = np.load(f).reshape(200, 100, -1)

with open('../test_0.npy', 'rb') as f:
    u0 = np.load(f).reshape(200, 100, -1)

u0.shape, e0.shape
```

```python
plt.imshow(u0)
```

```python
def plot_eu(e, u):
    # Example: generating some data of shape (200, 200, 1)
    data = u
    data_2d = data[:, :, 0]

    # Create coordinate arrays for x and y dimensions
    y = np.arange(data_2d.shape[0])
    x = np.arange(data_2d.shape[1])
    X1, Y1 = np.meshgrid(x, y)

    # Example: generating some data of shape (200, 200, 1)
    data2 = e
    # Convert data to 2D by taking the first channel
    data_2d2 = data2[:, :, 0]

    # Create coordinate arrays for x and y dimensions
    y = np.arange(data_2d2.shape[0])
    x = np.arange(data_2d2.shape[1])
    X2, Y2 = np.meshgrid(x, y)

    # Create a 3D figure and add a subplot with 3D projection
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(131, projection='3d')

    # Create the 3D contour plot
    contour = ax.contour3D(X1, Y1, data_2d, 50, cmap='viridis')  # 50 contour levels
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Data value')

    ax = fig.add_subplot(132, projection='3d')
    contour = ax.contour3D(X2, Y2, data_2d2, 50, cmap='viridis')  # 50 contour levels
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Data value')

    ax = fig.add_subplot(133, projection='3d')
    contour = ax.contour3D(X2, Y2, np.abs(data_2d2 - data_2d), 50, cmap='viridis')  # 50 contour levels
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Data value')

    plt.title('3D Contour Plot')
    plt.show()
```

```python
plot_eu(e0, u0)
```

```python
with open('../exact_45000.npy', 'rb') as f:
    e50 = np.load(f).reshape(200, 100, -1)

with open('../test_45000.npy', 'rb') as f:
    u50 = np.load(f).reshape(200, 100, -1)

plot_eu(e50, u50)
```
